```python
import numpy as np
import random
```

```python
#1.- FUNCIÓN TABLERO VACÍO

#creo función crear tablero donde podemos elegir el tamaño del mismo.
# por defecto le marcamos el tamaño de 10x10
#el caracter _ representa el agua

def crear_tablero(tamano= 10):

    tablero = np.full((tamano,tamano),"_") 
    print(tablero)
    return tablero

#tablero_yo = en el que juega el ordenador
#tablero_pc = en el que juego yo(tienen que no ser visibles los barcos)
tablero_yo = crear_tablero() # como en la función he especificado que tamano = 10, por defecto nuestro tablero es 10x10 
tablero_pc = crear_tablero()
```

```python
#2 MI TABLERO: FUNCIÓN PARA COLOCAR LOS BARCOS CORRECTAMENTE: (OJO QUE NO COLOCA EL BARCO, SOLO COMPRUEBA SI EXISTE ALGÚN ERROR)

def colocar_barco_yo(barco, tablero_yo):
    num_max_filas = tablero_yo.shape[0]                                            #definimos el número máximo de filas del tablero (en la matriz es el parámetro 0)
    num_max_columnas = tablero_yo.shape [1]                                        #definimos el número máximo de columnas del tablero (en la matriz es el parámetro 1)
    tablero_temp_yo = tablero_yo.copy() 
    
    for pieza in barco:                                                         #para cada pieza del barco:
        fila = pieza[0]                                                         #definimos que el primer parámetro de cada pieza es la fila en la que está
        columna = pieza[1]                                                      #y que el segundo parámetro de casa pieza es la columna en la que está
        
        if fila < 0 or fila >= num_max_filas:                                   #estamos acotando las filas del tablero
            print(f"No puedo poner la pieza {pieza} porque se sale del tablero")
            return False                                                         #nos devuelve False porque esa pieza quedaría fuera del tablero 
        
        if columna < 0 or columna >= num_max_columnas:                           #hacemos lo mismo pero con las columnas
            print(f"No puedo poner la pieza {pieza} porque se sale del tablero")
            return False
        
        if tablero_temp_yo[fila,columna] == "O" or tablero_temp_yo[fila, columna] == "X":   #si la pieza nueva la queremos colocar encima de una pieza de barco o encima de una pieza ya disparada
            print(f"No puedo poner la pieza {pieza} porque hay otro barco")     #nos sale el error de que en esa casilla ya existe una parte de barco
            return False
    for pieza in barco:                                                         #colocamos las piezas en el tablero
        fila = pieza [0]
        columna = pieza[1]
        tablero_yo[fila, columna] = "O"
    return tablero_yo           #si todas las piezas del barco están en agua, nos imprime el tablero
```

```python
#3.-MI TABLERO CREAR UN BARCO DE UNA DETERMINADA LONGITUD CON POSICIÓN ALEATORIA
def crear_barco_yo(eslora, tablero_yo):
    num_max_filas = tablero_yo.shape[0]
    num_max_columnas = tablero_yo.shape[1]
    while True:                                                 #bucle while True hasta que encuentre una posición correcta para el barco
        barco = []                                              #aquí se almacenan las coordenadas del barco cuando son correcta
        pieza_ini = (random.randint(0,num_max_filas-1), random.randint (0, num_max_columnas-1)) #pieza inicial aleatoria
        print("pieza inicial: ", pieza_ini)                                                     
        barco.append(pieza_ini)                                                                 #añadimos la pieza a la lista barco       
        orientacion = random.choice(["N","S", "E", "O"])                                        #orientacion aleatoria
        print("con orientación ", orientacion)
        
        fila = pieza_ini [0]        #definimos la variable fila, que es el primer indice de pieza_ini                                                      
        columna = pieza_ini [1]     #definimos la variable columna, que es el segundo indice de pieza_ini   

        for i in range(eslora-1):   #especificamos la longitud de la eslora y lo iteramos para que nos vaya devolviendo las piezas, también tiene encuenta la orientación elegida aleatoriamente
            if orientacion == "N":
                fila -= 1
            elif orientacion == "S":
                fila += 1
            elif orientacion == "E":
                columna += 1
            elif orientacion == "O":
                columna -= 1
            pieza = (fila, columna)
            barco.append(pieza)        #añade la nueva pieza a lista barco, así hasta que esté completada su longitud

        resultado = colocar_barco_yo(barco,tablero_yo.copy())    #almacenamos el resutado del primer barco en el tablero copia
        
        if isinstance(resultado,np.ndarray):                #comprueba si el barco es válido en el tablero y si es correcto, nos devuelve el tablero con el barco
            return resultado
        else:
            print("Tengo que intentar colocar otro barco")
            pass
        
```

```python
# 4 YO: FUNCIÓN COLOCAR LOS BARCOS ALEATORIOS

def colocar_barcos_yo(tablero_yo):
        for i in range(3): #para que recorra el bucle 3 veces porque queremos 3 barcos de eslora 2
            tablero_yo = crear_barco_yo(2, tablero_yo)

        for i in range(2): #para que recorra el bucle 2 veces porque queremos 2 barcos de eslora 3
            tablero_yo = crear_barco_yo(3, tablero_yo)

        for i in range(1): #para que recorra el bucle 1 vez porque queremos 1 barco de eslora 1
            tablero_yo = crear_barco_yo(4, tablero_yo)
    
        print(tablero_yo)
        return tablero_yo

tablero_yo = colocar_barcos_yo(tablero_yo) 
```

```python
#DISPAROS ALEATORIOS DE LA MÁQUINA TENIENDO EN CUENTA QUE SI EL RESULTADO ES TOCADO, REPITE EL TIRO (pero no es muy lista y el siguiente tiro lo coloca alteatorio y no al lado del anterior disparo)

def disparo_maquina(tablero_yo):
    num_max_filas = tablero_yo.shape[0]
    num_max_columnas = tablero_yo.shape[1]
    
    while True:         
      casilla = (random.randint(0,num_max_filas-1), random.randint (0, num_max_columnas-1))

      disparo_resultante = ""
    
      if tablero_yo[casilla] == "O":   #si hay barco
          tablero_yo[casilla] = "X"    #lo sustituimos por x
          print("Tocado")
          disparo_resultante = "Tocado"

      elif tablero_yo[casilla] == "X": #si ya ha disparado ahí , tenemos que configurar a la máquina para que no repita disparos
          print("Agonia, deja de perder el tiempo, dispara a otro sitio")
          disparo_resultante = "Repetido"  
      else:                           #si el hueco es agua, sustituimos el hueco por una A para que se vea que ya ha disparado ahí
          tablero_yo[casilla] = "A"
          print("Agua")
          disparo_resultante = "Agua"

      if disparo_resultante == "Tocado":   #aquí estamos indicando si le toca repetir disparo
        pass
      else:
        break
    
    if "O" in tablero_yo:
        print("Aún quedan barcos que hundir,sigue jungando")
    else:
        print("Fin de la partida, ¡has ganado!")
    return tablero_yo
    
resultado_pc = disparo_maquina(tablero_yo)
print(tablero_yo)
```

```python
#4: PC: FUNCIÓN COLOCAR LOS BARCOS ALEATORIOS

def colocar_barcos_pc (tablero_pc):
    for i in range(3): #para que recorra el bucle 3 veces porque queremos 3 barcos de eslora 2
        tablero_pc = crear_barco_pc(2, tablero_pc)

    for i in range(2): #para que recorra el bucle 2 veces porque queremos 2 barcos de eslora 3
        tablero_pc = crear_barco_pc(3, tablero_pc)

    for i in range(1): #para que recorra el bucle 1 vez porque queremos 1 barco de eslora 1
        tablero_pc = crear_barco_pc(4, tablero_pc)
    
    print("tablero listo")
    return tablero_pc
    
tablero_pc = colocar_barcos_pc(tablero_pc) 
```

```python
#ESTO ES LA FUNCION PARA DISPARAR YO
def disparar(casilla,tablero_pc):
    
    if tablero_pc[casilla] == "O":   #si hay barco
        tablero_pc[casilla] = "X"    #lo sustituimos por x
        print("Tocado")

    elif tablero_pc[casilla] == "X": #si ya ha disparado ahí , tenemos que configurar a la máquina para que no repita disparos
        print("Agonia, deja de perder el tiempo, dispara a otro sitio") 
    
    else:                           #si el hueco es agua, sustituimos el hueco por una A para que se vea que ya ha disparado ahí
        tablero_pc[casilla] = "A"
        print("Agua")
      
    return tablero_pc
    
```

```python
#FUNCIÓN INPUT MI TURNO PERO SIN REPETICIÓN SI TOCO UN BARCO. SI NO QUEDAN BARCOS QUE HUNDIR SALE EL MENSAJE DE FIN DE LA PARTIDA

def mi_turno(tablero_pc) 
    x= int(input ("Por favor, elija donde disparar: fila: "))
    y= int(input ("columna: "))
    disparar((x,y),tablero_pc)
    print(f":ese es el resultado de las coordenadas {x}, {y}")
    
    if "O" in tablero_pc:                                     
        print("Aún quedan barcos que hundir,sigue jungando")
    else:
        print("Fin de la partida, ¡has ganado!")
    
    return
```
