```python
import numpy as np
import random
```

```python
def crear_tablero(tamano= 10):

    tablero = np.full((tamano,tamano),"_") 
    print(tablero)
    return tablero


tablero_yo = crear_tablero()
tablero_pc = crear_tablero()
```

```python
def colocar_barco_yo(barco, tablero_yo):
    num_max_filas = tablero_yo.shape[0]
    num_max_columnas = tablero_yo.shape [1]
    tablero_temp_yo = tablero_yo.copy() 
    
    for pieza in barco:
        fila = pieza[0]
        columna = pieza[1]
        
        if fila < 0 or fila >= num_max_filas:
            print(f"No puedo poner la pieza {pieza} porque se sale del tablero")
            return False
        
        if columna < 0 or columna >= num_max_columnas:
            print(f"No puedo poner la pieza {pieza} porque se sale del tablero")
            return False
        
        if tablero_temp_yo[fila,columna] == "O" or tablero_temp_yo[fila, columna] == "X":
            print(f"No puedo poner la pieza {pieza} porque hay otro barco")
            return False
    for pieza in barco:
        fila = pieza [0]
        columna = pieza[1]
        tablero_yo[fila, columna] = "O"
    return tablero_yo
```

```python

def crear_barco_yo(eslora, tablero_yo):
    num_max_filas = tablero_yo.shape[0]
    num_max_columnas = tablero_yo.shape[1]
    while True:
        barco = []
        pieza_ini = (random.randint(0,num_max_filas-1), random.randint (0, num_max_columnas-1))
        print("pieza inicial: ", pieza_ini)                                                     
        barco.append(pieza_ini)
        orientacion = random.choice(["N","S", "E", "O"])
        print("con orientación ", orientacion)
        
        fila = pieza_ini [0]
        columna = pieza_ini [1]

        for i in range(eslora-1):
            if orientacion == "N":
                fila -= 1
            elif orientacion == "S":
                fila += 1
            elif orientacion == "E":
                columna += 1
            elif orientacion == "O":
                columna -= 1
            pieza = (fila, columna)
            barco.append(pieza)

        resultado = colocar_barco_yo(barco,tablero_yo.copy())
        
        if isinstance(resultado,np.ndarray):
            return resultado
        else:
            print("Tengo que intentar colocar otro barco")
            pass
        
```

```python
def crear_barco_pc(eslora, tablero_pc):
    num_max_filas = tablero_pc.shape[0]
    num_max_columnas = tablero_pc.shape[1]
    while True:
        barco = []
        pieza_ini = (random.randint(0,num_max_filas-1), random.randint (0, num_max_columnas-1))                                               
        print("pieza inicial: colocada")  
        barco.append(pieza_ini)
        orientacion = random.choice(["N","S", "E", "O"])
        print("con orientación desconocida")
        
        fila = pieza_ini [0]
        columna = pieza_ini [1]

        for i in range(eslora-1):
            if orientacion == "N":
                fila -= 1
            elif orientacion == "S":
                fila += 1
            elif orientacion == "E":
                columna += 1
            elif orientacion == "O":
                columna -= 1
            pieza = (fila, columna)
            barco.append(pieza)

        resultado = colocar_barco_pc(barco,tablero_pc.copy())

        if isinstance(resultado,np.ndarray):
            return resultado
        else:
            print("Tengo que intentar colocar otro barco")
            pass
```
