## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomesticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas ultimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas.
* Método que calcula la media de ventas de los ultimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
#1
#Creamos tienda, todas las tiendas son de tipo electrodomésticos y están abiertas por defecto
class Tienda:
  tipo = "Electrodomésticos"
  abierta = True

#Definimos los atributos propios de cada tienda
  def __init__(self,nombre,direccion,empleados,ventas_1,ventas_2,ventas_3):
    self.nombre = nombre
    self.direccion = direccion
    self.empleados = int(empleados)
    self.ventas = [ventas_1,ventas_2,ventas_3]

#Definimos todas las características de una tienda en concreto (los atributos de clase y los atributos propios de cada tienda)

  def caracteristicas(self):
   return "Nombre: " + self.nombre +", Tipo: " + self.tipo + ", Tienda abierta: " + str(self.abierta) + ", Dirección: " + self.direccion + ", Numero empleados: " + str(self.empleados) + ", Ventas: " + str(self.ventas)


#Método para calcular las ventas de todos los meses, que devuelva un numero con todas las ventas

  def ventas_trimestre(self):
    ventas_1=self.ventas[0]
    ventas_2=self.ventas[1]
    ventas_3=self.ventas[2]

    return str(ventas_1+ventas_2+ventas_3)

#Método que calcula la media de ventas de los ultimos meses, por empleado:

  def ventas_x_empleado(self):
    ventas_trimestre= int(self.ventas_trimestre())
    empleados= self.empleados
    return ventas_trimestre/empleados

#Método que devuelve en un string el nombre de la tienda, junto con su dirección.

  def directorio(self):
    return "Nombre tienda: " + self.nombre + ", Dirección: " + self.direccion

#Método que obtiene las ventas del último mes.

  def ventas_tienda_mes_3(self):
    ventas_3=self.ventas[2]
    return ventas_3

#Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

  def marketing(self,x):
    ventas_trimestre = float(self.ventas_trimestre())
    self.x = x
    if x <1000:
      ventas_trimestre=ventas_trimestre*1.2
    else:
      ventas_trimestre=ventas_trimestre*1.5
    return ventas_trimestre
```

```python
#2 Crear tres tiendas con datos inventados
#Tienda 1
electro1= Tienda("Electro 1","Avenida lavadora", 3, 10000,7000,30000)

#Tienda 2
electro2= Tienda("Electro 2","Calle Nevera", 5, 50000,20000,25000)

#Tienda 3
electro3= Tienda("Electro 3","Glorieta Aspirador", 8, 80000,45000,72000)
```

```python
#4. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle for

ventas_tercer_mes= 0

tiendas=[electro1,electro2,electro3]
for tienda in tiendas:
  ventas_tercer_mes = ventas_tercer_mes + tienda.ventas[2]

print (ventas_tercer_mes)
```

```python
#Método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.

#Copiamos lo que teníamos antes en class Perro:
class Perro:
  patas = 4
  orejas = 2
  ojos = 2
  velocidad = 0

  def __init__(self,nombre,raza,pelo_marron= True, sin_dueño= True):
    self.nombre = nombre
    self.raza = raza
    self.pelo_marron = str(pelo_marron)
    self.sin_dueño = str(sin_dueño)

#Añadimos el método andar. Queremos que el resultado sea la velocidad que llevaba (al inicio es 0) + el aumento de velocidad que queramos darle

  def andar(self,aumento_velocidad):
    self.aumento_velocidad= aumento_velocidad
    self.velocidad = self.velocidad + aumento_velocidad
    return self.nombre+ " lleva una velocidad de " + str(self.velocidad)

```

```python
# Necesita otro método (parar), donde pondremos la velocidad a 0.

#Copiamos lo que teníamos antes en class Perro:
class Perro:
  patas = 4
  orejas = 2
  ojos = 2
  velocidad = 0

  def __init__(self,nombre,raza,pelo_marron= True, sin_dueño= True):
    self.nombre = nombre
    self.raza = raza
    self.pelo_marron = str(pelo_marron)
    self.sin_dueño = str(sin_dueño)

  def andar(self,aumento_velocidad):
    self.aumento_velocidad= aumento_velocidad
    self.velocidad = self.velocidad + aumento_velocidad
    return self.nombre+ " lleva una velocidad de " + str(self.velocidad)

# Método parar. Reasignamos la velocidad a 0.

  def parar(self):
    self.velocidad = 0
    return self.nombre+ " está parado"
```
