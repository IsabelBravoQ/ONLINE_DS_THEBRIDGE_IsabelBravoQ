## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
#forma larga
import math
grados_texto = input ("¿cuantos radianes son los siguientes grados? ")
grados = float (grados_texto)
print ("radianes =", grados*(math.pi/180))
```

```python
#forma corta
import math
grados = float (input ("¿cuantos radianes son los siguientes grados? "))
print ("radianes = ", grados*(math.pi/180))
```

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
#forma larga
base_text = input ("define base ")
base = float (base_text)
altura_text = input ("define altura ")
altura = float (altura_text)
print ("area =" , base*altura )
```

```python
#forma corta
base = float (input ("define base "))
altura = float (input ("define altura "))
print ("area = ", base*altura)
```

## Ejercicio 5
